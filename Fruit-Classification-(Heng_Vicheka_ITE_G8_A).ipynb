# 4.Cell Codes

Import the necessary libraries

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.optimizers import RMSprop
```

Load and display the image:

```python
img = image.load_img("D:/tensorflow/training/orange/3_100.jpg")
```

```python
plt.imshow(img)
```

```python
cv2.imread("D:/tensorflow/training/orange/3_100.jpg")
cv2.imread("D:/tensorflow/training/orange/3_100.jpg").shape
```

Define the ImageDataGenerator for training and validation:

```python
# Create an ImageDataGenerator object and load the training dataset and validation_dataset.
train = ImageDataGenerator(rescale=1/255)
validation = ImageDataGenerator(rescale=1/255)
```

Create the train and validation datasets using flow_from_directory:

```python

train_dataset = train.flow_from_directory("D:/tensorflow/training/", target_size=(200, 200), batch_size=3, class_mode='binary')
validation_dataset = validation.flow_from_directory("D:/tensorflow/validation/", target_size=(200, 200), batch_size=3, class_mode='binary')
```

```python
# these 2 below code sentences are used to get the list of classes that the classification model is trained on.
# Print the class indices and classes of the training dataset.
train_dataset.class_indices
```

Define the model architecture using the Sequential API:

```python
# Create a sequential model and add convolutional layers.
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3, 3), activation='relu', input_shape=(200, 200, 3)),
    tf.keras.layers.MaxPool2D(2, 2),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPool2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPool2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

Compile the model with the desired loss, optimizer, and metrics:

```python
# Compile the model and train it.
model.compile(loss='binary_crossentropy',
              optimizer=RMSprop(lr=0.001),
              metrics=["accuracy"])
```

Train the model using the training dataset:

```python
model_fit = model.fit(train_dataset, steps_per_epoch=3, epochs=10, validation_data=validation_dataset)
```

Define the directory path for testing images:

```python
#this code is use for visualizing the results of the classification process. 
#The code loads the images from the D:/tensorflow/testing directory, resizes them to 200x200 pixels, 
#and then classifies them using the model. The results of the classification are then displayed in a visualization.
# Load the testing images and predict their labels.
dir_path = "D:/tensorflow/testing"
```

Iterate over the images in the directory and perform prediction:

```python
for i in os.listdir(dir_path):
    img = image.load_img(dir_path + '/' + i, target_size=(200, 200))
    plt.imshow(img)
    plt.show()
    
    X = image.img_to_array(img)
    X = np.expand_dims(X, axis=0)
    images = np.vstack([X])
    val = model.predict(images)
    if val == 0:
        print("This is apple.")
    elif val == 1:
        print("This is banana.")
    else:
        print("This is orange.")
```
